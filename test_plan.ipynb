```python
import polyline
import json

encoded = """
cydeiAdh...
"""  # tu polyline completa

# OJO: precision=6
pts = polyline.decode(encoded, precision=6)  # [(lat, lng), ...]

geojson = {
  "type": "FeatureCollection",
  "features": [{
    "type": "Feature",
    "properties": {},
    "geometry": {
      "type": "LineString",
      "coordinates": [[lat, lng] for lat, lng in pts]  # GeoJSON = [lon, lat]
    }
  }]
}

print(json.dumps(geojson))
```

```python
with open("geojson.geojson", "w") as f:
    json.dump(geojson, f)
```
